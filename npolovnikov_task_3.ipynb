**Задание 3**

В игре Plants & Gardens каждый месяц проводятся тематические события, `ограниченные по времени`. В них игроки могут получить уникальные предметы для сада и персонажей, дополнительные монеты или бонусы. Для получения награды требуется пройти ряд уровней за определенное время. С помощью каких метрик можно оценить результаты последнего прошедшего события?


Для оценки временного тематического события можно рассмотреть следующие метрики:

1) Partiicipation Rate ( количество игроков, стартовавших событие, к общему количеству активных игроков за период). Данная метрика важна, чтобы понять насколько событие привлекло внимание среди уже активных пользователей.

2) DAU(Daily Active Users)/ MAU (Monthly Active Users). Рост этих метрик после запуска события указывает на его успешность.

3) Conversion Rate (% игроков, совершивших покупки в игре во время события). По данной метрике можно проверить стимулирует ли событие монетизацию.

4) ARPU / ARPPU . Метрики так же показывают, как событие повлияло на доход.

5) Дневной Retention Rate - процент пользователей, которые возвращаются в игру после первого запуска. Повышение данной метрики после старта события будет говорить о его ценности для удержания игроков.

6) Churn Rate - процент игроков, которые перестают пользоваться игрой за период события. Рост этой метрики может говорить о недовольстве старых игроков новым событием.

7) Средняя длина игровой сессии - событие может как стимулировать игрока больше времени проводить в приложении, так и утомлять его. Длина игровой сессии - это основная метрика, которая показывает PMF , т е ценность приложения для игрока.

7) CSAT (Customer Satisfaction Score) -  насколько пользователи довольны игрой и проводимыми в ней событиями. Так же можно понять, готовы ли игроки рекомендовать приложение другим людям.

**Механика усложнения**

Предположим, в другом событии мы усложнили механику событий так, что при каждой неудачной попытке выполнения уровня игрок будет откатываться на несколько уровней назад. Изменится ли набор метрик оценки результата? Если да, то как?

Набор метрик **должен расшириться**, так как усложнение механики меняет поведение, мотивацию и прогрессию игроков, а следовательно и монетизацию приложения. Необходимо следить за дополнительными метриками.

Дополнительные метрики, на которые следует смотреть:
    1. Средний достигнутый уровень в событии. Если этот показатель низкий, то игроки, вероятно, теряют мотивацию проходить событие после нескольких откатов. Это может привести в конечном счете к оттоку пользователей.
    2. Drop-off уровень - на каких уровнях игроки бросают прохождение события или испытывают много откатов. Данная метрика помогает увидеть "узкие места в событии".
    3. Среднее число откатов в прогрессе на игрока. По этой метрике можно понять насколько сложное/легкое событие, стоит ли внести изменения в уровни.
